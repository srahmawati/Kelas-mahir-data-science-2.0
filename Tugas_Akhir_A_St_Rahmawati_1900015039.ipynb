## Pertanyaan Analisis

1. Buku Non fiksi yang memiliki reviews terbanyak?
2. Berdasrkan data Amazon bestselling book, dari semua harga buku manakah yang mendapatkan ulasan tertinggi?

## Data Preparation

```
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
```

```
#mengimport file csv yang sudah di upload di github
df = pd.read_csv('https://raw.githubusercontent.com/srahmawati/Pusat-Studi-Data-Sains-UAD/main/bestsellers%20with%20categories.csv')
df
```

```
#untuk mengetahui jumlah data yang terdapat pada kolom Prise
df.Price.value_counts()
```

```
List_Price = list(pd.unique(df['Price']))
List_Price
```

```
list_sum = []
for i in List_Price:
  df_Rev = df[df['Price'] == i]
  sum_Rev = df_Rev.Reviews.sum()
  list_sum.append(sum_Rev)

print(list_sum)
```

```
# 1.menentukan Genre mana dengan reviews terbanyak!
df_reviews = df[df['Genre']=='Non Fiction']
df_reviews
```

```
#2. membuat data frame dari jumlah reviews pada setiap Price
df_Reviews = pd.DataFrame({"Price":List_Price,
              "Jumlah_Reviews":list_sum})
df_Reviews.sort_values(['Jumlah_Reviews', 'Price'], ascending=[True, False], inplace=True)
df_Reviews
```

## Data Visualization

```
#Membuat Plot 
gbr_amazon = px.bar(df_Reviews,
               x = 'Price',
               y = 'Jumlah_Reviews')

gbr_amazon.update_layout(
    title = 'Diagram Garis jumlah Reviews pada price tahun 2009 sampai 2019',
    title_font_color = 'blue',
    title_font_size = 30,

    xaxis = dict(
        title = 'Price',
        titlefont_size = 16,
        tickfont_size = 9),
    
    yaxis = dict(
        title = 'Jumlah Reviews',
        titlefont_size = 16,
        tickfont_size = 14))

gbr_amazon.show()
```
